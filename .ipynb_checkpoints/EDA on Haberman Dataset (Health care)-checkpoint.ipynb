3. Number of positive axillary nodes detected (numerical)  --> A positive axillary node is a lymph node in the area of the armpit (axilla) to which cancer has spread. This spread is determined by surgically removing some of the lymph nodes and examining them under a microscope to see whether cancer cells are present.
4. Survival status (class attribute) 
          -- 1 = the patient survived 5 years or longer 
          -- 2 = the patient died within 5 year

```python
#Status_1 = 1 -- patient survived more that 5 years
#Status_2 = 2 --patient survived less than 5 years

import numpy as np
status_1=hmds.loc[hmds["status"] == 1];
status_2=hmds.loc[hmds["status"] == 2];
plt.plot(status_1["age"],np.zeros_like(status_1["age"]),'o')
plt.plot(status_2["age"],np.zeros_like(status_2["age"]),'o')
plt.show()

#As we can see the 1D scatter plot of status_1 and status_2 are not clear; hence we have to try other options 
```

```python
#pairplots
plt.close();
sns.set_style("whitegrid");
sns.pairplot(hmds,hue='status',height=3);
plt.show();
```

```python
sns.FacetGrid(hmds,hue="status",size=5).map(sns.distplot,"age").add_legend();
plt.show();

#trying to classify with age would be a terrible idea
```

```python
#trying to classify with age would be a terrible idea; so lets try with other features

sns.FacetGrid(hmds,hue="status",size=5).map(sns.distplot,"nodes").add_legend();
sns.FacetGrid(hmds,hue="status",size=5).map(sns.distplot,"year").add_legend();
plt.show();
```

As we can observe; the comparision with nodes for status_1 and status_2 makes more sense
*People with less nodes has higher chances of living, we can observe that very clearly

NODES>>AGE>YEAR

```python
counts, bin_edges = np.histogram(hmds['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(hmds['nodes'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();
```
